```python
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\Vaishnavi\OneDrive\Desktop\project final\customer_shopping_data.csv")
data.info()
data.isnull().sum()
data['invoice_date']=pd.to_datetime(data['invoice_date'], format='%d-%m-%Y', errors='coerce')

#here  we created new column 
label=['Young Adult','Adult','Middle-Aged','Senior']

data["Aged_Group"]=pd.qcut(data["age"],q=4,labels=label)

data["Sales"]=data["price"] * data["quantity"]

data.columns=data.columns.str.title()

pip install psycopg2-binary sqlalchemy

from sqlalchemy import create_engine

# Step 1: Connection details
username = "postgres"
password = "root"      # replace with your password
host = "localhost"
port = "5432"
database = "customer_shopping_data"

# Step 2: Create connection engine
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 3: Load DataFrame into PostgreSQL
table_name = "customer"   # you can change table name

data.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
